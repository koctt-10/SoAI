Цель работы: Приобретение навыков, необходимых для начала работы с нейросетевым расширением.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = 8
xmin = 3
xmax = 10
dx = 0.07

def f(x):
    return -2*x*np.exp(-(x-3)**2) + 3*x*np.exp(-(x-5)**2)

y_1 = f(x_1)
```

```python
print(x_1, y_1)
```

```python
x = np.arange(xmin, xmax + dx, dx)
y = f(x)
```

```python
print(x , y)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = -2*x*np.exp(-(x-3)**2) + 3*x*np.exp(-(x-5)**2)') 
plt.scatter([x], [y], color='red', zorder=5)

# Оформление графика
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('График функции', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
